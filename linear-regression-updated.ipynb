```python
def plotter_2p(p1,p2):
    """Plots two points, and the line joining them.

    Args:
        p1: A tuple of two numbers representing the coordinates of the first point.
        p2: A tuple of two numbers representing the coordinates of the second point.
    """
    fig , axes = plt.subplots(2,2, figsize=(14,10))
    
    x1,y1 = p1
    x2,y2 = p2
    m = (y2-y1)/(x2-x1)
    b = y1 - m*x1
    
    axes[0,0].plot([0,0],[-10,25],color='black')
    axes[0,0].plot([-10,25],[0,0],color='black')
    axes[0,0].scatter([x1,x2],[y1,y2],color='orange')
    axes[0,0].set_xlim(0,10)
    axes[0,0].set_ylim(0,10)
    axes[0,0].set_title(f'$P_1({x1},{y1}),$ $P_2({x2},{y2})$')

    axes[0,1].plot([0,0],[-10,25],color='black')
    axes[0,1].plot([-10,25],[0,0],color='black')
    axes[0,1].plot([x1,x1],[y1,y2],linestyle='--',color='cyan')
    axes[0,1].plot([x1,x2],[y2,y2],linestyle='--',color='cyan')
    axes[0,1].plot([x1,x2],[y1,y2])
    axes[0,1].scatter([x1],[y2],marker='x',color='green')
    axes[0,1].scatter([x1,x2],[y1,y2],color='orange')
    axes[0,1].set_xlim(0,10)
    axes[0,1].set_ylim(0,10)
    axes[0,1].set_title(f'$d_x = |x_1 - x_2| = |{x1}-{x2}|={x2-x1}$\n$d_y = |y_1 - y_2| = |{y1}-{y2}|={y1-y2}$')

    axes[1,0].plot([0,0],[-10,25],color='black')
    axes[1,0].plot([-10,25],[0,0],color='black')
    axes[1,0].plot([x1,x1],[y1,y2],linestyle='--',color='cyan')
    axes[1,0].plot([x1,x2],[y2,y2],linestyle='--',color='cyan')
    axes[1,0].scatter([x1],[y2],marker='x',color='green')
    axes[1,0].scatter([x1,x2],[y1,y2],color='orange')
    axes[1,0].plot([x1,x2],[y1,y2])
    axes[1,0].plot([-5,15],[-m*5+b,m*15+b],linestyle='--',color='magenta')
    axes[1,0].plot([-5,15],[b,b],linestyle='--',color='green')
    axes[1,0].set_xlim(-5,15)
    axes[1,0].set_ylim(-5,15)
    axes[1,0].set_title(f'$m = (y_2 - y_1)/(x_2 - x_1) = {m}$\n$b = y - mx = {b}$')
    
    axes[1,1].plot([0,0],[-10,25],color='black')
    axes[1,1].plot([-10,25],[0,0],color='black')
    axes[1,1].plot([-10,25],[-m*10+b,m*25+b],color='magenta')
    axes[1,1].plot([-10,25],[b,b],linestyle='--',color='green')
    axes[1,1].set_xlim(-10,25)
    axes[1,1].set_ylim(-10,25)
    axes[1,1].set_title('Final Line')
    
    plt.suptitle('Line of an Equation from Cartesian Coordinates of Two different points.')
    plt.savefig(str(p1+p2)+'.jpg')
```

```python
def plot_vertical_line(ax, x, y1, y2, linestyle='--', color='blue'):
    ax.plot([x, x], [y1, y2], linestyle=linestyle, color=color)

def plot_horizontal_line(ax, y, x1, x2, linestyle='--', color='blue'):
    ax.plot([x1, x2], [y, y], linestyle=linestyle, color=color)


def plotter_2p_1(p1, p2):
    """Plots two points, and the line joining them.

    Args:
        p1: A tuple of two numbers representing the coordinates of the first point.
        p2: A tuple of two numbers representing the coordinates of the second point.
    """
    fig, axes = plt.subplots(2, 2, figsize=(12, 12))

    x1, y1 = p1
    x2, y2 = p2
    m = (y2 - y1) / (x2 - x1)
    b = y1 - m * x1

    [ax.set_xlim(-10, 10) and ax.set_ylim(-10, 10) and ax.plot([0, 0], [-10, 10], color='black') and ax.plot([-10, 10], [0, 0], color='black') for ax in axes.flat]
    [ax.scatter([x1, x2], [y1, y2], color='blue') for ax in axes.flat]

    # Plot dashed lines connecting the points
    [plot_vertical_line(ax, x1, y1, y2) for ax in axes[(0,1),(1,0)]]
    [plot_horizontal_line(ax, y2, x1, x2) for ax in axes[(0,1),(1,0)]]
    axes[1, 0].plot([-10, 10], [-m * 10 + b, m * 10 + b], color='magenta')
    axes[1, 0].plot([-10, 10], [b, b], color='green')

    # Plot the final line
    axes[1, 0].plot([x1, x2], [y1, y2], color='orange')
    axes[1, 1].plot([-10, 10], [-m * 10 + b, m * 10 + b], color='magenta')
    axes[1, 1].plot([-10, 10], [b, b], color='green')

    # Set titles
    axes[0, 0].set_title(f'$P_1({x1},{y1}),$ $P_2({x2},{y2})$')
    axes[0, 1].set_title(f'$d_x = |x_1 - x_2| = |{x1}-{x2}|={x2 - x1}$\n$d_y = |y_1 - y_2| = |{y1}-{y2}|={y1 - y2}$')
    axes[1, 0].set_title(f'$m = (y_2 - y_1)/(x_2 - x_1) = ({y2} - {y1})/({x2} - {x1}) = {m}$\n$b = y_i - mx_i =  = {y1} - {m} * {x1}={b}$')
    axes[1, 1].set_title('Final Line')

    # Set super title
    plt.suptitle('Line of an Equation from Cartesian Coordinates of Two different points.')

    # Save and show plot
    plt.savefig(str(p1 + p2) + '.jpg')
    plt.show()
```

```python
def simple_data_generator(size = 100, slope = 37.1802, left = 5, right = 15, noise = 0):
    np.random.seed(42)
    independent = np.random.randint(left-2,right+2,size)
    dependent = slope * independent + np.random.randint(left-noise,right+noise,size)
    return pd.DataFrame({"X" : independent, "y": dependent})

data = simple_data_generator(20,15,5,100,40)
sns.scatterplot(x=data["X"],y=data["y"])
# for i in range(data.shape[0]):
    
# plt.scatter(data['X'],np.zeros(data.shape[0]))
# plt.scatter(np.zeros(data.shape[0]),data['y'])
```

```python
class OLS:
    def __init__(self):
        self.m_ = 0
        self.b_ = 0
        self.E_ = 0

    def fit(self, X_train, y_train):
        """
        This function takes x and y as input and return
        m and b calculated using Ordinary Least Squared method.

        """
        assert len(X_train) == len(y_train)
        x_bar = np.mean(X_train)
        y_bar = np.mean(y_train)
        self.m_ = np.sum((X_train - x_bar) * (y_train - y_bar)) / np.sum((X_train - x_bar) ** 2)
        self.b_ = y_bar - self.m_ * x_bar

    def predict(self, X_test):
        y_pred = self.m_ * X_test + self.b_
        return y_pred

    def average_error(self, y_test, y_pred):
        assert len(y_test) == len(y_pred)
        self.E_ = np.mean((y_test - y_pred) ** 2)
        return self.E_

    def get_coeff_intercept(self):
        return self.m_, self.b_

    def plot_regression_line(self, X_train, y_train):
        """Plots the regression line over a scatter plot of the training data.

        Args:
            X_train: A numpy array of the training data, with each row representing a
                single data point and each column representing a single feature.
            y_train: A numpy array of the target values for the training data.
        """

        plt.figure()
        plt.scatter(X_train, y_train, color='blue')

        # Plot the regression line
        x_min = np.min(X_train)
        x_max = np.max(X_train)
        y_min = self.m_ * x_min + self.b_
        y_max = self.m_ * x_max + self.b_
        plt.plot([x_min, x_max], [y_min, y_max], color='magenta')

        plt.xlabel('X')
        plt.ylabel('Y')
        plt.title(f'Regression Line (coeff: {self.m_}, intercept: {self.b_})')
        plt.show()


ols = OLS()
ols.fit(np.array(X_train).reshape(-1, 1), np.array(y_train).reshape(-1, 1))
print(ols.get_coeff_intercept())
y_simple_preds = ols.predict(X_test)
print(ols.average_error(np.array(y_test), np.array(y_simple_preds)))

# Plot the regression line
ols.plot_regression_line(X_train, y_train)
```

```python
# Regression Class

class LinReg:
    
    def __init__(self):
        self.coef_ = None
        self.intercept_ = None
        self.mae = None
        self.mse = None
        self.rmse = None
        self.r2 = None
    def fit(self, X_train, y_train):
        X_train = np.insert(X_train, 0, 1, axis=1)

        betas = np.linalg.inv(X_train.T@X_train)@X_train.T@y_train
        self.intercept_=betas[0]
        self.coef_=betas[1:]
        
    def predict(self, X_test):
        y_pred = X_test@self.coef_+self.intercept_
        return y_pred
    
    def score(self, y_test, y_pred):
        self.mae = mean_absolute_error(y_test, y_pred)
        self.mse = mean_squared_error(y_test, y_pred)
        self.rmse = np.sqrt(self.mse)
        self.r2 = r2_score(y_test, y_pred)
        
    def __str__(self):
        return f"Coefficient: {self.coef_}\nIntercept: {self.intercept_}\n\nMAE: {self.mae}\nMSE: {self.mse}\nRMSE: {self.rmse}\nR2: {self.r2}"

        
```
